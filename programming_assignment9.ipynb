# 1.	Write a Python program to check if the given number is a Disarium Number?

```python
num = int(input("Enter a no. :"))
```

```python
def check_disarium(num):
    import math
    s = str(num)
    res = 0
    for i in range(len(s)):
        res = res + math.pow(int(s[i]),i+1)
    print("Original number :",num)
    print("Disarium number :",int(res))
    if(int(res) == num):
        print("Number is a disarium")
    else:
        print("Number is not a disarium")
```

```python
check_disarium(num)
```

# 3.	Write a Python program to check if the given number is Happy Number?

```python
def check_happy_No():
    import math
    num = int(input("Enter a number : "))
    s = str(num)
    res = 0
    while num != 1:
        for i in range(len(s)):
            res = int(res + math.pow(int(s[i]),2))
        print(res)
        num = res
        s = str(num)
        res = 0
    if num == 1:
        return "It is happy number"
```

```python
check_happy_No()
```

# 5.	Write a Python program to determine whether the given number is a Harshad Number?

```python
def harshad_no():
    num = int(input("Enter a number"))
    s = str(num)
    sum = 0
    for i in range(len(s)):
        sum = sum + int(s[i])
    if num % sum == 0:
        print("Number is Harshad number")
    else:
        print("Number is not harshad number")
```

```python
harshad_no()
```
